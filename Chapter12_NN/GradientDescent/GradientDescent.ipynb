# Gradient Descent

<font size="4">

$x = x - \eta \cdot \nabla f(x)$  
$f(x)$ ist die Funktion (z.B. Fehlerfunktion)  
$x$ ist der Parameter (z.B. Gewicht im NN)  

</font>

```python
def plot_data(x: np.ndarray, y: np.ndarray) -> None:
    plt.plot(x, y, color="blue")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.show()
```

```python
def plot_training(x: np.ndarray, y: np.ndarray, p: Union[float, List[float]]) -> None:
    if isinstance(p, list):
        for index, point in enumerate(p):
            plt.scatter(point, f(point), color="green", marker="o", s=50)
    else:
        plt.scatter(p, f(p), color="green", marker="o", s=200)
    plt.scatter(0, f(0), color="red", marker="*", s=200)
    plt.plot(x, y, color="blue")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.show()
```

```python
def f(x):
    return x**2

def f_prime(x):
    return 2 * x
```

```python
x = np.arange(-20, 20, 0.1)
y = [f(val) for val in x]
```

```python
plot_data(x, y)
```

```python
p = np.random.uniform(-20, 20)

plot_training(x, y, p)
```

```python
eta = 1e-3

x_current = np.random.uniform(-20, 20)
y_current = f(x_current)

stop_conv = 1e-6
stop_div = 1e+6
stop_iter = 100000
it = 0

downhill_points = []

while y_current > stop_conv and y_current < stop_div and it < stop_iter:
    x_current = x_current - eta * f_prime(x_current)

    it += 1
    y_current = f(x_current)
    if it % 100 == 0:
        downhill_points.append(x_current)

print("Solution: ", y_current)
print("X = ", x_current)
```

```python
plot_training(x, y, downhill_points)
```

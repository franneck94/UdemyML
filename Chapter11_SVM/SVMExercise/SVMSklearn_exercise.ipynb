```python
# Plot random images
for index, (image, label) in enumerate(zip(x[:10], y[:10])):
    plt.subplot(2, 5, index + 1)
    plt.axis('off')
    plt.imshow(image.reshape((8, 8)), cmap=plt.cm.gray_r, interpolation='nearest')
    plt.title(f'Train: {label + 1}')
```

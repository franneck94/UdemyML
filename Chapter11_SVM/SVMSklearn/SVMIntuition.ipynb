#### Kernel Trick

```python
def quadratic_kernel(x: np.ndarray) -> np.ndarray:
    x[:, 0] = x[:, 0] ** 2
    x[:, 1] = x[:, 1] ** 2
    return x
```

```python
x_qk = quadratic_kernel(x)
```

```python
plt.scatter(x_qk[:, 0], x_qk[:, 1], c=y)
plt.xlabel("$x_1$")
plt.ylabel("$x_2$")
plt.xlim(-1, 20)
plt.ylim(-1, 20)
plt.show()
```

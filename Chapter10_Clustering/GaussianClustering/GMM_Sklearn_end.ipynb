```python
import numpy as np


np.random.seed(42)
```

```python
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture

from dataset import generate_dataset
from plotting import plot_results
```

```python
x, y = generate_dataset(50_000, 50_000)

idxs = np.random.permutation(len(x))

x = x[idxs]
y = y[idxs]
```

```python
plt.scatter(x[:1000, 0], x[:1000, 1], c=y[:1000])
plt.show()
```

```python
n_components = 2

gauss = GaussianMixture(n_components=n_components, covariance_type="diag")
gauss.fit(x)
```

```python
print(f"Model converged: {gauss.converged_}")
```

```python
covs = gauss.covariances_
means = gauss.means_

cov1 = np.diag(covs[0])
cov2 = np.diag(covs[1])
```

```python
print(f"Cov 1. Distribution:\n{cov1}")
print(f"Cov 2. Distribution:\n{cov2}")
print(f"Mean 1. Distribution:\n{means[0]}")
print(f"Mean 2. Distribution:\n{means[1]}")
```

```python
plot_results(
    X=x[:1000],
    Y=y[:1000],
    Y_=gauss.predict(x[:1000]),
    means=means,
    covariances=[cov1, cov2],
    index=0,
    title="GM",
)
```

```python
plot_results(
    X=x[:1000],
    Y=y[:1000],
    Y_=y[:1000],
    means=means,
    covariances=[cov1, cov2],
    index=0,
    title="GM",
)
```

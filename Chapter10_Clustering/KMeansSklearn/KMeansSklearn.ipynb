```python
import numpy as np


np.random.seed(42)
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

from plotting import plot
```

```python
dataset = load_iris()

x = dataset.data[:, [0, 1]]
y = dataset.target
```

```python
colors = ["red", "green", "blue"]

for idx, point in enumerate(x):
    plt.scatter(point[0], point[1], color=colors[y[idx]])
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3, max_iter=1_000, n_init=10)
kmeans.fit(x)

random_idxs = np.random.randint(low=0, high=len(x), size=(40))
x_test = x[random_idxs]
y_test = y[random_idxs]

y_pred = kmeans.predict(x_test).astype(int)

print(f"cluster_centers_:\n{kmeans.cluster_centers_}")
print(f"y_pred:\n{y_pred}")
```

```python
plot(x_test, y_pred, kmeans.cluster_centers_, kmeans)
```

Objective: $[0, inf)$  
Score: $(-inf, 0]$  

```python
score = kmeans.score(x, y)

print(f"Score: {score}")
```
